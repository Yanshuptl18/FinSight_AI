## 24. Build Keyword Lookup

```python
def keyword_sector_scores(text):

    scores = {}

    if pd.isna(text):
        return scores

    text = str(text).lower()

    for sector, keywords in SECTOR_KEYWORDS.items():

        count = 0

        for keyword in keywords:

            if keyword.lower() in text:

                count += 1

        if count > 0:

            scores[sector] = count

    return scores
```

## 25. Event Score

```python
def event_sector_score(event):

    if pd.isna(event):
        return {}

    sector = EVENT_TO_SECTOR.get(event)

    if sector is None:
        return {}

    return {

        sector: 2.5

    }
```

## 26. Topic Score

```python
def topic_sector_score(topic):

    if pd.isna(topic):

        return {}

    for prefix, sector in TOPIC_TO_SECTOR.items():

        if topic.startswith(prefix):

            return {

                sector: 2.0

            }

    return {}
```

## 27. Entity Score

```python
def entity_sector_score(entity_list):

    scores = {}

    if not isinstance(entity_list, list):

        return scores

    for entity in entity_list:

        entity = str(entity).lower()

        for sector, keywords in SECTOR_KEYWORDS.items():

            if entity in [

                k.lower()

                for k in keywords

            ]:

                scores[sector] = (

                    scores.get(sector, 0)

                    + 1.5

                )

    return scores
```

## 28. Hybrid Scoring Function

```python
def combine_sector_scores(row):

    final_scores = {}

    headline_scores = keyword_sector_scores(

        row["headline"]

    )

    event_scores = event_sector_score(

        row["final_event"]

    )

    topic_scores = topic_sector_score(

        row["topic_name"]

    )

    entity_scores = entity_sector_score(

        row["entities"]

    )

    for source in [

        headline_scores,

        event_scores,

        topic_scores,

        entity_scores

    ]:

        for sector, score in source.items():

            final_scores[sector] = (

                final_scores.get(sector, 0)

                + score

            )

    return final_scores
```

## 29. Calculate Scores

```python
tqdm.pandas()

financial_df["sector_scores"] = (

    financial_df

    .progress_apply(

        combine_sector_scores,

        axis=1

    )

)
```

## 30. Assign Primary Sector

```python
def choose_sector(score_dict):

    if len(score_dict) == 0:

        return "Other"

    return max(

        score_dict,

        key=score_dict.get

    )
```

```python
financial_df["sector"] = (

    financial_df["sector_scores"]

    .apply(choose_sector)

)
```

## 31. Confidence

```python
def sector_confidence(scores):

    if len(scores) == 0:

        return 0

    total = sum(scores.values())

    best = max(scores.values())

    return round(best / total, 3)
```

```python
financial_df["sector_confidence"] = (

    financial_df["sector_scores"]

    .apply(

        sector_confidence

    )

)
```

# 5. Sector Risk Intelligence Engine

## 43. Company Sector Distribution

```python
company_sector_distribution = (

    financial_df

    .groupby(

        [

            "ticker",

            "sector"

        ]

    )

    .size()

    .reset_index(

        name="news_count"

    )

)

display(company_sector_distribution.head())
```

## 45. Merge company analytics.

## 46. Weighted Sector Intelligence.

```python
sector_risk = (

    company_sector_distribution

    .groupby("sector")

    .agg(

        companies=(

            "ticker",

            "nunique"

        ),

        weighted_company_risk=(

            "normalized_risk_score",

            lambda x: np.average(

                x,

                weights=

                company_sector_distribution.loc[

                    x.index,

                    "sector_share"

                ]

            )

        ),

        weighted_company_score=(

            "composite_score",

            lambda x: np.average(

                x,

                weights=

                company_sector_distribution.loc[

                    x.index,

                    "sector_share"

                ]

            )

        ),

        weighted_market_stability=(

            "market_stability_score",

            lambda x: np.average(

                x,

                weights=

                company_sector_distribution.loc[

                    x.index,

                    "sector_share"

                ]

            )

        )

    )

    .reset_index()

)

display(sector_risk)
```

## 51. Sector Risk Fingerprint

```python
def build_sector_risk_fingerprint(row):

    tags = []

    if row["sector_risk_level"] in ["Very Low", "Low"]:
        tags.append("Low Risk")

    elif row["sector_risk_level"] == "Medium":
        tags.append("Moderate Risk")

    else:
        tags.append("High Risk")

    if row["market_stability_category"] == "Very Stable":
        tags.append("Stable Market")

    elif row["market_stability_category"] == "Stable":
        tags.append("Balanced Market")

    if row["company_strength"] in ["Strong", "Excellent"]:
        tags.append("Strong Companies")

    if row["event_diversity"] >= 20:
        tags.append("Highly Diversified")

    if row["coverage_score"] >= 100:
        tags.append("High Coverage")

    return ", ".join(tags)


sector_profile["sector_risk_fingerprint"] = (

    sector_profile

    .apply(

        build_sector_risk_fingerprint,

        axis=1

    )

)

display(

    sector_profile[

        [

            "sector",

            "sector_risk_fingerprint"

        ]

    ].head(10)

)
```

# 6. Sector Relationship & Knowledge Graph

## 54. Build Sector × Company Matrix

```python
sector_company_matrix = (

    financial_df

    .pivot_table(

        index="sector",

        columns="ticker",

        values="news_id",

        aggfunc="count",

        fill_value=0

    )

)

print(sector_company_matrix.shape)

display(

    sector_company_matrix.iloc[:5, :5]

)
```

## 55. Sector Similarity Matrix

```python
from sklearn.metrics.pairwise import cosine_similarity

sector_similarity = pd.DataFrame(

    cosine_similarity(

        sector_company_matrix

    ),

    index=sector_company_matrix.index,

    columns=sector_company_matrix.index

)

print(sector_similarity.shape)

display(

    sector_similarity.iloc[:5,:5]

)
```

## 56. Find Strong Sector Relationships

```python
sector_edges = []

threshold = 0.55

for i in range(len(sector_similarity)):

    for j in range(i+1,len(sector_similarity)):

        sim = sector_similarity.iloc[i,j]

        if sim >= threshold:

            sector_edges.append({

                "sector_1":sector_similarity.index[i],

                "sector_2":sector_similarity.index[j],

                "similarity":round(sim,4)

            })

sector_edges = pd.DataFrame(

    sector_edges

)

print(sector_edges.shape)

display(

    sector_edges.head(20)

)
```

## 57. Shared Company Count

```python
company_sets = {}

for sector in financial_df["sector"].unique():

    company_sets[sector] = set(

        financial_df.loc[

            financial_df["sector"]==sector,

            "ticker"

        ].unique()

    )
```

```python
shared_companies = []

for s1 in company_sets:

    for s2 in company_sets:

        if s1 >= s2:

            continue

        overlap = len(

            company_sets[s1]

            &

            company_sets[s2]

        )

        if overlap>0:

            shared_companies.append({

                "sector_1":s1,

                "sector_2":s2,

                "shared_companies":overlap

            })

shared_companies = pd.DataFrame(

    shared_companies

)

display(

    shared_companies.head()

)
```

## 58. Merge Similarity + Shared Companies

```python
sector_network = (

    sector_edges

    .merge(

        shared_companies,

        on=[

            "sector_1",

            "sector_2"

        ],

        how="left"

    )

)

sector_network["shared_companies"] = (

    sector_network["shared_companies"]

    .fillna(0)

    .astype(int)

)

display(

    sector_network.head(20)

)
```

## 59. Build the Graph

```python
import networkx as nx

sector_graph = nx.Graph()

for _, row in sector_network.iterrows():

    sector_graph.add_edge(

        row["sector_1"],

        row["sector_2"],

        similarity=row["similarity"],

        shared_companies=row["shared_companies"]

    )

print("Nodes :", sector_graph.number_of_nodes())

print("Edges :", sector_graph.number_of_edges())
```

## 61. Betweenness Centrality

```python
betweenness = nx.betweenness_centrality(

    sector_graph,

    weight="similarity"

)

betweenness = pd.DataFrame({

    "sector": betweenness.keys(),

    "betweenness": betweenness.values()

})

betweenness = betweenness.sort_values(

    "betweenness",

    ascending=False

)

display(betweenness)
```

## 62. Closeness Centrality

```python
closeness = nx.closeness_centrality(

    sector_graph

)

closeness = pd.DataFrame({

    "sector": closeness.keys(),

    "closeness": closeness.values()

})

closeness = closeness.sort_values(

    "closeness",

    ascending=False

)

display(closeness)
```

## 63. Merge All Centralities

```python
sector_network_metrics = (

    degree

    .merge(

        betweenness,

        on="sector"

    )

    .merge(

        closeness,

        on="sector"

    )

)

display(

    sector_network_metrics

)
```

# 7. Sector Temporal Intelligence Engine

## 71. Create Monthly Sector Timeline

```python
sector_timeline = (

    financial_df

    .groupby(

        [

            "sector",

            "year",

            "month"

        ]

    )

    .agg(

        news_count=("news_id", "count"),

        companies=("ticker", "nunique"),

        publishers=("publisher", "nunique"),

        avg_confidence=("sector_confidence", "mean")

    )

    .reset_index()

)

display(
    sector_timeline.head(20)
)
```

## 72. Create Date Column

```python
sector_timeline["date"] = pd.to_datetime(

    dict(

        year=sector_timeline["year"],

        month=sector_timeline["month"],

        day=1

    )

)

display(

    sector_timeline.head()

)
```

## 73. Sort Chronologically

```python
sector_timeline = (

    sector_timeline

    .sort_values(

        [

            "sector",

            "date"

        ]

    )

    .reset_index(drop=True)

)

display(

    sector_timeline.head(20)

)
```

## 74. Monthly Growth Rate

```python
sector_timeline["previous_news"] = (

    sector_timeline

    .groupby("sector")["news_count"]

    .shift(1)

)

sector_timeline["growth_rate"] = (

    (

        sector_timeline["news_count"]

        -

        sector_timeline["previous_news"]

    )

    /

    sector_timeline["previous_news"]

)

sector_timeline["growth_rate"] = (

    sector_timeline["growth_rate"]

    .replace(

        [

            np.inf,

            -np.inf

        ],

        np.nan

    )

)

display(

    sector_timeline.head(20)

)
```

## 75. Rolling Average (3 Months)

```python
sector_timeline["rolling_news"] = (

    sector_timeline

    .groupby("sector")["news_count"]

    .transform(

        lambda x:

        x.rolling(

            3,

            min_periods=1

        ).mean()

    )

)

display(

    sector_timeline.head(20)

)
```

## 76. Calculate Momentum Score

```python
sector_timeline["previous_rolling"] = (

    sector_timeline

    .groupby("sector")["rolling_news"]

    .shift(1)

)

sector_timeline["momentum_score"] = (

    sector_timeline["rolling_news"]

    -

    sector_timeline["previous_rolling"]

)

sector_timeline["momentum_score"] = (

    sector_timeline["momentum_score"]

    .fillna(0)

)

display(

    sector_timeline[

        [

            "sector",

            "date",

            "rolling_news",

            "momentum_score"

        ]

    ].head(20)

)
```

## 77. Detect Trend Direction

```python
def trend_direction(x):

    if x > 0.25:

        return "Growing"

    elif x < -0.25:

        return "Declining"

    else:

        return "Stable"


sector_timeline["trend"] = (

    sector_timeline["momentum_score"]

    .apply(trend_direction)

)

display(

    sector_timeline[

        [

            "sector",

            "date",

            "news_count",

            "rolling_news",

            "momentum_score",

            "trend"

        ]

    ].head(20)

)
```

### 78. Peak Activity Detection

```python
sector_peak = (

    sector_timeline

    .loc[

        sector_timeline

        .groupby("sector")["news_count"]

        .idxmax()

    ]

    .reset_index(drop=True)

)

display(

    sector_peak[

        [

            "sector",

            "date",

            "news_count"

        ]

    ].head(20)

)
```

## 79. Overall Temporal Statistics

```python
sector_temporal = (

    sector_timeline

    .groupby("sector")

    .agg(

        first_month=("date","min"),

        latest_month=("date","max"),

        active_months=("date","count"),

        average_news=("news_count","mean"),

        max_news=("news_count","max"),

        average_growth=("growth_rate","mean"),

        average_momentum=("momentum_score","mean")

    )

    .reset_index()

)

display(

    sector_temporal.head(20)

)
```

## 80. Merge Peak Information

```python
sector_temporal = (

    sector_temporal

    .merge(

        sector_peak[

            [

                "sector",

                "date",

                "news_count"

            ]

        ]

        .rename(

            columns={

                "date":"peak_month",

                "news_count":"peak_news"

            }

        ),

        on="sector",

        how="left"

    )

)

display(

    sector_temporal.head(20)

)
```

## 81. Assign Lifecycle Stage

```python
def sector_lifecycle(row):

    if row["average_growth"] > 0.30:

        return "Emerging"

    elif row["average_growth"] > 0.10:

        return "Growing"

    elif row["average_growth"] > -0.10:

        return "Mature"

    else:

        return "Declining"


sector_temporal["lifecycle"] = (

    sector_temporal

    .apply(

        sector_lifecycle,

        axis=1

    )

)

display(

    sector_temporal[

        [

            "sector",

            "average_growth",

            "lifecycle"

        ]

    ]

)
```

## 82. Trend Strength Score

```python
sector_temporal["trend_strength"] = (

    sector_temporal["average_growth"].abs()

    *

    np.log1p(

        sector_temporal["average_news"]

    )

)

display(

    sector_temporal[

        [

            "sector",

            "trend_strength"

        ]

    ]

)
```

## 83. Temporal Intelligence Score

```python
sector_temporal["temporal_score"] = (

      sector_temporal["average_news"] * 0.35

    + sector_temporal["trend_strength"] * 0.30

    + sector_temporal["average_momentum"] * 15 * 0.20

    + sector_profile.set_index("sector")

        .loc[

            sector_temporal["sector"],

            "avg_confidence"

        ]

        .values

        *100

        *0.15

)

display(

    sector_temporal[

        [

            "sector",

            "temporal_score"

        ]

    ]

    .sort_values(

        "temporal_score",

        ascending=False

    )

)
```

## 84. Temporal Ranking

## 85. Seasonality Score

```python
seasonality = (

    sector_timeline

    .groupby("sector")

    .agg(

        monthly_std=(

            "news_count",

            "std"

        ),

        monthly_mean=(

            "news_count",

            "mean"

        )

    )

    .reset_index()

)

seasonality["seasonality_score"] = (

    seasonality["monthly_std"]

    /

    seasonality["monthly_mean"]

)

display(

    seasonality.head()

)
```

## 87. Temporal Fingerprint

```python
def temporal_fingerprint(row):

    tags = []

    if row["lifecycle"] == "Emerging":

        tags.append("Fast Growing")

    elif row["lifecycle"] == "Growing":

        tags.append("Growing")

    elif row["lifecycle"] == "Mature":

        tags.append("Mature")

    else:

        tags.append("Declining")

    if row["seasonality_score"] > 1:

        tags.append("Highly Seasonal")

    elif row["seasonality_score"] > 0.50:

        tags.append("Seasonal")

    else:

        tags.append("Consistent")

    if row["average_news"] > 1000:

        tags.append("High Activity")

    elif row["average_news"] > 300:

        tags.append("Medium Activity")

    else:

        tags.append("Low Activity")

    return ", ".join(tags)


sector_temporal["temporal_fingerprint"] = (

    sector_temporal

    .apply(

        temporal_fingerprint,

        axis=1

    )

)

display(

    sector_temporal[

        [

            "sector",

            "temporal_fingerprint"

        ]

    ]

)
```

## 88. Temporal Knowledge Card

## 104. Confidence Category

```python
def confidence_category(score):

    if score >= 80:

        return "Very High"

    elif score >= 65:

        return "High"

    elif score >= 50:

        return "Medium"

    else:

        return "Low"


sector_recommendations["confidence_level"] = (

    sector_recommendations["recommendation_confidence"]

    .apply(

        confidence_category

    )

)

display(

    sector_recommendations[

        [

            "sector_1",

            "recommendation_confidence",

            "confidence_level"

        ]

    ]

)
```

## 105. Diversification Category

```python
def diversification_level(count):

    if count >= 6:

        return "Highly Diversified"

    elif count >= 4:

        return "Diversified"

    elif count >= 2:

        return "Moderately Diversified"

    else:

        return "Focused"


sector_recommendations["diversification_level"] = (

    sector_recommendations["recommendation_count"]

    .apply(

        diversification_level

    )

)

display(

    sector_recommendations[

        [

            "sector_1",

            "recommendation_count",

            "diversification_level"

        ]

    ]

)
```

## 106. Recommendation Explanation

```python
sector_recommendations["recommendation_summary"] = (

    "Recommendation: "

    + sector_recommendations["recommendation"]

    + " | "

    + "Confidence: "

    + sector_recommendations["confidence_level"]

    + " | "

    + "Diversification: "

    + sector_recommendations["diversification_level"]

)

display(

    sector_recommendations[

        [

            "sector_1",

            "recommendation_summary"

        ]

    ].head(20)

)
```

## 107. Investment Theme

```python
def investment_theme(row):

    if (

        row["recommendation"] == "Strong Buy"

        and

        row["sector_risk_level"] == "Low"

    ):

        return "Long Term Compounder"

    elif row["recommendation"] == "Buy":

        return "Growth Opportunity"

    elif row["recommendation"] == "Watch":

        return "Monitor"

    else:

        return "Speculative"


sector_recommendations["investment_theme"] = (

    sector_recommendations

    .apply(

        investment_theme,

        axis=1

    )

)

display(

    sector_recommendations[

        [

            "sector_1",

            "investment_theme"

        ]

    ]

)
```

## 108. Recommendation Strength

```python
sector_recommendations["recommendation_strength"] = (

      sector_recommendations["opportunity_score"] * 0.60

    + sector_recommendations["recommendation_confidence"] * 0.40

)

sector_recommendations["recommendation_strength"] = (

    sector_recommendations["recommendation_strength"]

    .round(2)

)

display(

    sector_recommendations[

        [

            "sector_1",

            "recommendation_strength"

        ]

    ]

)
```

## 109. Final Recommendation Rank

## 129. Build Search Function

```python
def search_sector(query):

    query = str(query).lower().strip()

    matched = []

    for keyword, sectors in keyword_index.items():

        if query in keyword:

            matched.extend(sectors)

    matched = sorted(

        list(set(matched))

    )

    return matched
```
