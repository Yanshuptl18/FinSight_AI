## 1. Imports

```python
import os
import warnings

import numpy as np
import pandas as pd

warnings.filterwarnings("ignore")
```

## 2. Paths

```python
PROJECT_PATH = "/content/drive/MyDrive/FinSight_AI"

DATA_PATH = os.path.join(
    PROJECT_PATH,
    "data"
)

PROCESSED_PATH = os.path.join(
    DATA_PATH,
    "processed"
)

INTELLIGENCE_PATH = os.path.join(
    PROCESSED_PATH,
    "financial_intelligence"
)

PROFILE_PATH = os.path.join(
    PROCESSED_PATH,
    "company_profiles"
)

os.makedirs(
    PROFILE_PATH,
    exist_ok=True
)
```

## 3. Load Financial Intelligence Dataset

```python
intelligence_df = pd.read_parquet(

    os.path.join(

        INTELLIGENCE_PATH,

        "financial_intelligence_dataset.parquet"

    )

)


print("Financial Intelligence Dataset")


print(intelligence_df.shape)

intelligence_df.head()
```

## 4. Dataset Statistics

```python
print("Dataset Statistics")

print(f"Total Headlines : {len(intelligence_df):,}")

print(f"Companies : {intelligence_df['ticker'].nunique():,}")

print(f"Publishers : {intelligence_df['publisher'].nunique():,}")

print(f"Events : {intelligence_df['final_event'].nunique():,}")

print(f"Years : {intelligence_df['year'].nunique():,}")
```

## 5. Company Basic Statistics

```python
company_basic = (

    intelligence_df

    .groupby("ticker")

    .agg(

        total_news=(

            "news_id",

            "count"

        ),

        first_news=(

            "published_date",

            "min"

        ),

        latest_news=(

            "published_date",

            "max"

        ),

        unique_publishers=(

            "publisher",

            "nunique"

        ),

        avg_headline_length=(

            "headline_length",

            "mean"

        ),

        avg_word_count=(

            "word_count",

            "mean"

        )

    )

    .reset_index()

)

company_basic.head()
```

## 6. Event Statistics

```python
# Event Statistics
event_stats = (

    pd.crosstab(

        intelligence_df["ticker"],

        intelligence_df["final_event"]

    )

    .reset_index()

)

event_stats.head()
```

## 7. Market Signal Statistics

```python
# Market Signal Statistics
market_stats = (

    pd.crosstab(

        intelligence_df["ticker"],

        intelligence_df["market_signal"]

    )

    .reset_index()

)

market_stats.head()
```

## 8. Convert Market Counts into Percentages

```python
# Market Signal Percentages
market_percentage = market_stats.copy()

signals = [

    col

    for col in market_percentage.columns

    if col != "ticker"

]

total_signal = market_percentage[signals].sum(axis=1)

for col in signals:

    market_percentage[col + "_pct"] = (

        market_percentage[col]

        / total_signal

        * 100

    ).round(2)

market_percentage.head()
```

## 9. Confidence Statistics

```python
# Confidence Statistics
confidence_stats = (

    intelligence_df

    .groupby("ticker")

    .agg(

        avg_confidence=(

            "final_confidence",

            "mean"

        ),

        median_confidence=(

            "final_confidence",

            "median"

        ),

        min_confidence=(

            "final_confidence",

            "min"

        ),

        max_confidence=(

            "final_confidence",

            "max"

        ),

        confidence_std=(

            "final_confidence",

            "std"

        )

    )

    .reset_index()

)

confidence_stats.head()
```

## 10. Top Publisher Statistics

```python
# Publisher Statistics
publisher_stats = (

    intelligence_df

    .groupby(

        [

            "ticker",

            "publisher"

        ]

    )

    .size()

    .reset_index(

        name="headline_count"

    )

)

top_publishers = (

    publisher_stats

    .sort_values(

        [

            "ticker",

            "headline_count"

        ],

        ascending=[

            True,

            False

        ]

    )

    .groupby("ticker")

    .first()

    .reset_index()

)

top_publishers = top_publishers.rename(

    columns={

        "publisher":"top_publisher",

        "headline_count":"top_publisher_news"

    }

)

top_publishers.head()
```

## 11. Activity Statistics

```python
# Activity Statistics
activity_stats = (

    intelligence_df

    .groupby("ticker")

    .agg(

        active_years=(

            "year",

            "nunique"

        ),

        active_months=(

            "month",

            "nunique"

        ),

        active_weekdays=(

            "weekday",

            "nunique"

        ),

        average_news_per_year=(

            "year",

            lambda x: len(x) / x.nunique()

        )

    )

    .reset_index()

)

activity_stats.head()
```

## 12. Company Risk Score

```python
# Company Risk Score
RISK_WEIGHTS = {

    "Fraud":8,

    "Bankruptcy":10,

    "Cybersecurity":6,

    "Litigation":5,

    "Layoffs":4,

    "Credit Rating":3,

    "Regulatory Approval":-2,

    "Dividend":-2,

    "Earnings":-1,

    "Share Buyback":-2

}

risk_score = event_stats.copy()

risk_score["risk_score"] = 0

for event, weight in RISK_WEIGHTS.items():

    if event in risk_score.columns:

        risk_score["risk_score"] += (

            risk_score[event]

            * weight

        )

risk_score = risk_score[

    [

        "ticker",

        "risk_score"

    ]

]

risk_score.head()
```

## 13. Intelligence Score

```python
# Company Intelligence Score
intelligence_score = (

    company_basic

    [

        [

            "ticker",

            "total_news",

            "unique_publishers"

        ]

    ]

    .merge(

        confidence_stats[

            [

                "ticker",

                "avg_confidence"

            ]

        ],

        on="ticker"

    )

)

intelligence_score["intelligence_score"] = (

    np.log1p(

        intelligence_score["total_news"]

    )

    *

    intelligence_score["avg_confidence"]

    *

    np.log1p(

        intelligence_score["unique_publishers"]

    )

)

intelligence_score = intelligence_score[

    [

        "ticker",

        "intelligence_score"

    ]

]

intelligence_score.head()
```

## 14. Merge All Company Statistics

```python
# Merge All Company Statistics
company_profiles = company_basic.copy()

company_profiles = company_profiles.merge(

    confidence_stats,

    on="ticker",

    how="left"

)

company_profiles = company_profiles.merge(

    top_publishers,

    on="ticker",

    how="left"

)

company_profiles = company_profiles.merge(

    activity_stats,

    on="ticker",

    how="left"

)

company_profiles = company_profiles.merge(

    risk_score,

    on="ticker",

    how="left"

)

company_profiles = company_profiles.merge(

    intelligence_score,

    on="ticker",

    how="left"

)

company_profiles = company_profiles.merge(

    market_percentage,

    on="ticker",

    how="left"

)

company_profiles = company_profiles.merge(

    event_stats,

    on="ticker",

    how="left"

)


print("Company Profile Created")
print(company_profiles.shape)

company_profiles.head()
```

## 15. Fill Missing Values

```python
# Fill Missing Values

numeric_columns = company_profiles.select_dtypes(

    include=["number"]

).columns

company_profiles[numeric_columns] = company_profiles[

    numeric_columns

].fillna(0)

company_profiles.head()
```

## 26. Rank Companies

```python
# Company Rankings
company_profiles["risk_rank"] = (

    company_profiles["risk_score"]

    .rank(

        ascending=False,

        method="dense"

    )

    .astype(int)

)

company_profiles["intelligence_rank"] = (

    company_profiles["intelligence_score"]

    .rank(

        ascending=False,

        method="dense"

    )

    .astype(int)

)

company_profiles.head()
```

## 17. Save Company Profiles

```python
# Save Company Profiles
OUTPUT_FILE = os.path.join(

    PROFILE_PATH,

    "company_profiles.parquet"

)

company_profiles.to_parquet(

    OUTPUT_FILE,

    index=False

)

print("COMPANY PROFILES SAVED")

print(OUTPUT_FILE)
```
