## 1. Imports

```python
import os
import warnings

import numpy as np
import pandas as pd

warnings.filterwarnings("ignore")
```

## 2. Paths

```python
PROJECT_PATH = "/content/drive/MyDrive/FinSight_AI"

DATA_PATH = os.path.join(
    PROJECT_PATH,
    "data"
)

PROCESSED_PATH = os.path.join(
    DATA_PATH,
    "processed"
)

INTELLIGENCE_PATH = os.path.join(
    PROCESSED_PATH,
    "financial_intelligence"
)

TIMELINE_PATH = os.path.join(
    PROCESSED_PATH,
    "company_timeline"
)

os.makedirs(
    TIMELINE_PATH,
    exist_ok=True
)
```

## 3. Load Financial Intelligence Dataset

```python
timeline_df = pd.read_parquet(

    os.path.join(

        INTELLIGENCE_PATH,

        "financial_intelligence_dataset.parquet"

    )

)

print("Financial Intelligence Dataset")

print(timeline_df.shape)

timeline_df.head()
```

## 4. Select Timeline Columns

```python
timeline_df = timeline_df[

    [

        "news_id",

        "ticker",

        "published_date",

        "headline",

        "publisher",

        "final_event",

        "market_signal",

        "final_confidence"

    ]

].copy()

timeline_df.head()
```

## 5. Data Preparation

```python
timeline_df["published_date"] = pd.to_datetime(

    timeline_df["published_date"]

)

timeline_df = timeline_df.sort_values(

    [

        "ticker",

        "published_date"

    ]

).reset_index(

    drop=True

)

print("Timeline Prepared")

timeline_df.head()
```

## 6. Event Sequence Number

```python
# Event Sequence
timeline_df["event_number"] = (

    timeline_df

    .groupby("ticker")

    .cumcount()

    + 1

)

timeline_df.head()
```

## 7. Previous & Next Event Dates

```python
# Previous & Next Event
timeline_df["previous_event_date"] = (

    timeline_df

    .groupby("ticker")["published_date"]

    .shift(1)

)

timeline_df["next_event_date"] = (

    timeline_df

    .groupby("ticker")["published_date"]

    .shift(-1)

)

timeline_df.head()
```

## 8. Days Between Events

```python
# Time Gap Features
timeline_df["days_since_previous_event"] = (

    timeline_df["published_date"]

    -

    timeline_df["previous_event_date"]

).dt.days

timeline_df["days_until_next_event"] = (

    timeline_df["next_event_date"]

    -

    timeline_df["published_date"]

).dt.days

timeline_df.head()
```

## 9. Rolling Event Count

```python
# Rolling Total Events
timeline_df["events_so_far"] = (

    timeline_df

    .groupby("ticker")

    .cumcount()

    + 1

)

timeline_df.head()
```

## 10. Rolling Risk Events

```python
# Risk Events
RISK_EVENTS = [

    "Fraud",

    "Bankruptcy",

    "Cybersecurity",

    "Litigation",

    "Layoffs"

]

timeline_df["risk_event"] = (

    timeline_df["final_event"]

    .isin(RISK_EVENTS)

).astype(int)

timeline_df["risk_events_so_far"] = (

    timeline_df

    .groupby("ticker")["risk_event"]

    .cumsum()

)

timeline_df.head()
```

## 11. Rolling Bullish / Bearish Counts

```python
# Rolling Market Signals
timeline_df["bullish_event"] = (

    timeline_df["market_signal"]

    == "Bullish"

).astype(int)

timeline_df["bearish_event"] = (

    timeline_df["market_signal"]

    == "Bearish"

).astype(int)

timeline_df["bullish_events_so_far"] = (

    timeline_df

    .groupby("ticker")["bullish_event"]

    .cumsum()

)

timeline_df["bearish_events_so_far"] = (

    timeline_df

    .groupby("ticker")["bearish_event"]

    .cumsum()

)

timeline_df.head()
```

## 12. Event Importance Score

```python
# Event Importance Score
EVENT_IMPORTANCE = {

    "Bankruptcy": 10,
    "Fraud": 10,
    "Merger & Acquisition": 9,
    "Earnings": 9,
    "Regulatory Approval": 8,
    "Cybersecurity": 8,
    "Monetary Policy": 8,
    "Layoffs": 7,
    "Share Buyback": 7,
    "Product Launch": 7,
    "Revenue Guidance": 7,
    "Dividend": 6,
    "Funding": 6,
    "IPO": 6,
    "Litigation": 6,
    "Credit Rating": 5,
    "Analyst Rating": 5,
    "Executive Change": 5,
    "Commodity": 4,
    "Economic Data": 4,
    "Partnership": 4,
    "Stock Split": 4,
    "Insider Trading": 4,
    "Cryptocurrency": 3,
    "Market Movement": 3,
    "ESG": 3,
    "Options Activity": 2,
    "Technical Analysis": 2,
    "Trading Halt": 2,
    "Other": 1

}

timeline_df["event_importance"] = (

    timeline_df["final_event"]

    .map(EVENT_IMPORTANCE)

    .fillna(1)

    .astype(int)

)

timeline_df.head()
```

## 13. Event Severity Level

```python
# Event Severity
def get_event_severity(score):

    if score >= 9:
        return "Critical"

    elif score >= 7:
        return "High"

    elif score >= 4:
        return "Medium"

    else:
        return "Low"


timeline_df["event_severity"] = (

    timeline_df["event_importance"]

    .apply(get_event_severity)

)

timeline_df.head()
```

## 14. Daily Event Count

```python
# Daily Event Count
timeline_df["event_date"] = (

    timeline_df["published_date"]

    .dt.date

)

timeline_df["daily_event_count"] = (

    timeline_df

    .groupby(

        [

            "ticker",

            "event_date"

        ]

    )["news_id"]

    .transform("count")

)

timeline_df.head()
```

## 15. Company Activity Level

```python
# Company Activity Level
def get_activity_level(count):

    if count >= 20:
        return "Very High"

    elif count >= 10:
        return "High"

    elif count >= 5:
        return "Medium"

    elif count >= 2:
        return "Low"

    else:
        return "Very Low"


timeline_df["activity_level"] = (

    timeline_df["daily_event_count"]

    .apply(get_activity_level)

)

timeline_df.head()
```

## 16. Is Latest Event

```python
# Latest Event Flag
latest_dates = (

    timeline_df

    .groupby("ticker")["published_date"]

    .transform("max")

)

timeline_df["is_latest_event"] = (

    timeline_df["published_date"]

    == latest_dates

)

timeline_df.head()
```

## 17. Save Company Event Timeline

```python
# Save Company Event Timeline
OUTPUT_FILE = os.path.join(

    TIMELINE_PATH,

    "company_event_timeline.parquet"

)

timeline_df.to_parquet(

    OUTPUT_FILE,

    index=False

)

print("COMPANY EVENT TIMELINE SAVED")
print(OUTPUT_FILE)
```

## 18. Event Distribution

```python
# Event Distribution
event_summary = (

    timeline_df

    .groupby("final_event")

    .agg(

        headlines=("news_id", "count"),

        avg_confidence=("final_confidence", "mean")

    )

    .sort_values(

        "headlines",

        ascending=False

    )

    .reset_index()

)

display(event_summary)
```

## 19. Severity Distribution

```python
# Event Severity Distribution

severity_summary = (

    timeline_df

    .groupby("event_severity")

    .agg(

        headlines=("news_id", "count")

    )

    .sort_values(

        "headlines",

        ascending=False

    )

    .reset_index()

)

display(severity_summary)
```

## 20. Market Signal Distribution

```python
# Market Signal Distribution
signal_summary = (

    timeline_df

    .groupby("market_signal")

    .agg(

        headlines=("news_id", "count"),

        avg_confidence=("final_confidence", "mean")

    )

    .sort_values(

        "headlines",

        ascending=False

    )

    .reset_index()

)

display(signal_summary)
```

## 21. Most Active Companies

```python
# Most Active Companies
top_companies = (

    timeline_df

    .groupby("ticker")

    .agg(

        total_events=("news_id", "count"),

        latest_event=("published_date", "max"),

        avg_confidence=("final_confidence", "mean")

    )

    .sort_values(

        "total_events",

        ascending=False

    )

    .head(20)

    .reset_index()

)

display(top_companies)
```

## 22. Highest Importance Events

```python
# Highest Importance Events
important_events = (

    timeline_df

    .sort_values(

        [

            "event_importance",

            "final_confidence"

        ],

        ascending=[

            False,

            False

        ]

    )

    [

        [

            "ticker",

            "published_date",

            "headline",

            "final_event",

            "event_importance",

            "event_severity",

            "market_signal",

            "final_confidence"

        ]

    ]

    .head(20)

)

display(important_events)
```

## 23. Timeline Statistics

```python
# Timeline Statistics
timeline_statistics = pd.DataFrame({

    "Metric":[

        "Total Timeline Events",

        "Companies",

        "Publishers",

        "Unique Events",

        "Critical Events",

        "High Events",

        "Average Confidence",

        "Output File"

    ],

    "Value":[

        len(timeline_df),

        timeline_df["ticker"].nunique(),

        timeline_df["publisher"].nunique(),

        timeline_df["final_event"].nunique(),

        (timeline_df["event_severity"]=="Critical").sum(),

        (timeline_df["event_severity"]=="High").sum(),

        round(

            timeline_df["final_confidence"].mean(),

            4

        ),

        os.path.basename(

            OUTPUT_FILE

        )

    ]

})

display(timeline_statistics)
```
