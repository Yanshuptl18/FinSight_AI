## 1. Imports

```python
import os
import warnings

import pandas as pd

warnings.filterwarnings("ignore")
```

## 2. Paths

```python
PROJECT_PATH = "/content/drive/MyDrive/FinSight_AI"

DATA_PATH = os.path.join(
    PROJECT_PATH,
    "data"
)

PROCESSED_PATH = os.path.join(
    DATA_PATH,
    "processed"
)

CLASSIFICATION_PATH = os.path.join(
    DATA_PATH,
    "classification"
)

HYBRID_PATH = os.path.join(
    CLASSIFICATION_PATH,
    "hybrid_classification"
)

INTELLIGENCE_PATH = os.path.join(
    PROCESSED_PATH,
    "financial_intelligence"
)

os.makedirs(
    INTELLIGENCE_PATH,
    exist_ok=True
)
```

## 3. Load Financial News

```python
financial_news = pd.read_parquet(

    os.path.join(

        PROCESSED_PATH,

        "financial_news_clean.parquet"

    )

)

print("="*70)
print("Financial News")
print("="*70)

print(financial_news.shape)

financial_news.head()
```

## 4. Load Hybrid Classification

```python
classification = pd.read_parquet(

    os.path.join(

        HYBRID_PATH,

        "news_classification_final.parquet"

    )

)

print("Hybrid Classification")

print(classification.shape)

classification.head()
```

## 5. Validate Merge Keys

```python
print("Validation")

print("Financial News")

print("Rows :", len(financial_news))

print("Unique news_id :", financial_news["news_id"].nunique())

print()

print("Classification")

print("Rows :", len(classification))

print("Unique news_id :", classification["news_id"].nunique())
```

## 6. Merge Financial News and Classification

```python
intelligence_df = financial_news.merge(

    classification[

        [

            "news_id",

            "final_event",

            "market_signal",

            "final_confidence",

            "classification_source"

        ]

    ],

    on="news_id",

    how="left",

    validate="one_to_one"

)

print("Financial Intelligence Dataset Created")

print(intelligence_df.shape)

intelligence_df.head()
```

## 7. Data Validation

```python
print("Validation")

print(f"Rows                : {len(intelligence_df):,}")

print(f"Unique News IDs     : {intelligence_df['news_id'].nunique():,}")

print()

print("Missing Values")

print()

print(

    intelligence_df[

        [

            "final_event",

            "market_signal",

            "final_confidence",

            "classification_source"

        ]

    ]

    .isna()

    .sum()

)
```

## 8. Save Final Intelligence Dataset

```python
OUTPUT_FILE = os.path.join(

    INTELLIGENCE_PATH,

    "financial_intelligence_dataset.parquet"

)

intelligence_df.to_parquet(

    OUTPUT_FILE,

    index=False

)

print("FINANCIAL INTELLIGENCE DATASET SAVED")

print(OUTPUT_FILE)
```

## 9. Evaluation

```python
event_summary = (

    intelligence_df

    .groupby("final_event")

    .agg(

        headlines=("news_id", "count"),

        avg_confidence=("final_confidence", "mean")

    )

    .sort_values(

        "headlines",

        ascending=False

    )

    .reset_index()

)

display(event_summary)
```

## 10. Summary

```python
summary = pd.DataFrame({

    "Metric": [

        "Rows",

        "Columns",

        "Companies",

        "Publishers",

        "Event Types",

        "Output File"

    ],

    "Value": [

        len(intelligence_df),

        intelligence_df.shape[1],

        intelligence_df["ticker"].nunique(),

        intelligence_df["publisher"].nunique(),

        intelligence_df["final_event"].nunique(),

        os.path.basename(OUTPUT_FILE)

    ]

})

display(summary)
```
