## 2. Import Libraries

```python
import os
import warnings

import numpy as np
import pandas as pd

warnings.filterwarnings("ignore")

pd.set_option("display.max_columns", None)
pd.set_option("display.max_colwidth", 200)
```

## 3. Load Project Paths

```python
PROJECT_ROOT = "/content/drive/MyDrive/FinSight_AI"

RAW_DATA = os.path.join(PROJECT_ROOT, "data/raw")
```

## 4. Detect Available CSV Files

```python
csv_files = sorted([
    file
    for file in os.listdir(RAW_DATA)
    if file.endswith(".csv")
])

print("CSV Files Found")
for i, file in enumerate(csv_files, start=1):
    print(f"{i}. {file}")
```

## 5. Create Dataset Inspection Function

```python
def inspect_dataset(file_path):

    print(f"Dataset : {os.path.basename(file_path)}")

    df = pd.read_csv(file_path, low_memory=False)

    print("\nShape")
    print(df.shape)

    print("\nMemory Usage")
    print(f"{df.memory_usage(deep=True).sum()/1024**2:.2f} MB")

    print("\nColumns")
    print(df.columns.tolist())

    print("\nData Types")
    print(df.dtypes)

    print("\nFirst Five Rows")
    display(df.head())

    print("\nLast Five Rows")
    display(df.tail())

    print("\nMissing Values")
    display(
        df.isnull().sum()
        .sort_values(ascending=False)
    )

    print("\nDuplicate Rows")

    print(df.duplicated().sum())

    print("\nUnique Values")

    display(df.nunique())

    return df
```

## 6. Inspect Every Dataset

```python
datasets = {}

for file in csv_files:

    path = os.path.join(RAW_DATA, file)

    datasets[file] = inspect_dataset(path)
```

## 7. Dataset Summary Table

```python
summary = []

for name, df in datasets.items():

    summary.append({

        "Dataset": name,

        "Rows": df.shape[0],

        "Columns": df.shape[1],

        "Memory (MB)": round(
            df.memory_usage(deep=True).sum()/1024**2,
            2
        ),

        "Missing Cells": int(df.isnull().sum().sum()),

        "Duplicate Rows": int(df.duplicated().sum())

    })

summary_df = pd.DataFrame(summary)

summary_df
```

## 8. Compare Column Names

```python
print("COLUMN COMPARISON")
for name, df in datasets.items():

    print(f"\n{name}")

    print("-"*60)

    for col in df.columns:

        print(col)
```

## 9. Date Column Inspection

```python
print("DATE COLUMN CHECK")
for name, df in datasets.items():

    print(f"\n{name}")

    date_columns = [

        col

        for col in df.columns

        if "date" in col.lower()

        or "time" in col.lower()

    ]

    print(date_columns)
```

## 10. Company/Ticker Column Inspection

```python
print("COMPANY / TICKER COLUMNS")
for name, df in datasets.items():

    print(f"\n{name}")

    cols = [

        col

        for col in df.columns

        if "ticker" in col.lower()

        or "stock" in col.lower()

        or "company" in col.lower()

    ]

    print(cols)
```

## 11. Text Column Inspection

```python
print("TEXT COLUMNS")

for name, df in datasets.items():

    print(f"\n{name}")

    cols = [

        col

        for col in df.columns

        if "headline" in col.lower()

        or "title" in col.lower()

        or "text" in col.lower()

    ]

    print(cols)
```
